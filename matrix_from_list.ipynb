```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('data_clinical_patient_clean.txt', delimiter='\t')
data.head()
```

```python
data.head()
series = data['THERAPIES_XRC_LIST']
patients = list(data['PATIENT_ID'])
```

```python
patients_therapies = list(series)
```

```python
patients
```

```python
therapies = []
for patient in patients_therapies:
    try:
        therapies += patient.split(',')
    except:
        pass
```

```python
unique_thers =  (list(set(therapies)))
```

```python
len(patients)
```

```python
len(patients_therapies)
```

```python
ther_map = {}
for i, ther in enumerate(unique_thers):
    ther_map[ther] = []
    
    
```

```python
ther_map
```

```python
for i,patient in enumerate(patients_therapies):
    try:
        therapies = patient.split(',')
        for key in ther_map:
            if key in therapies:
                ther_map[key].append(1)
            else:
                ther_map[key].append(0)
    except:
        for key in ther_map:
            ther_map[key].append(0)
ther_map['PATIENT_ID'] = patients
```

```python
ther_map
```

```python
pd.DataFrame(ther_map).to_csv('clean_therapies.csv', sep=',')
```

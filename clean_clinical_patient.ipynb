```python
import numpy as np
import pandas as pd
```

```python
# I have been putting not done as the same value as unknown, but there may be merit in keeping the distinction aka there is a erasong why they did not do the test
# However, I am not too sure about what to put for Unknown as opposed to Not Done (they can't both be NaN)

ONCOTREE_CODE = {'BREAST':0,'IDC':1, 'MDLC':2, 'ILC':3}
CALC_TREATMENT_NAIVE = {'YES':1,'NO':0}
CALC_MET_SETTING = {'METASTATIC DISEASE PRESENT':1, 'NO METASTATIC DISEASE PRESENT':0, 'UNKNOWN':np.NaN}
BX_DCIS = {'Yes':1, 'No':0}
BX_LCIS = {'Yes':1, 'No':0}
BX_GRADE = {'Unknown':np.NaN, 'II Intermediate Grade (Moderately Differentiated)':2,
       'III High Grade (Poorly Differentiated)':3, 'Not Done':np.NaN, 'I-II':1.5,
       'I Low Grade (Well Differentiated)':1, 'II-III':2.5}
BX_ER = {'Positive':1, 'Unknown':np.NaN, 'Negative':-1, 'Not Done':np.NaN}
BX_PR = {'Positive':1, 'Unknown':np.NaN, 'Negative':-1, 'Not Done':np.NaN}
BX_HER2OVERALL = {'Unknown':np.NaN, 'Negative':-1, 'Not Done':np.NaN, 'Positive':1, 'Equivocal':0,
       'Equivicol':0}
BX_HER2IHC = {'Unknown':np.NaN, '0':0, '1+':1, 'Not Done':np.NaN, '3+':3, '2+':2, 'Negative':-1} 
BX_HER2FISH = {'Unknown':np.NaN, 'Negative':-1, 'Not Done':np.NaN, 'Positive':1, 'Not done':np.NaN}
BXHER2_RATIO = {'Unknown':np.NaN,'1':1,'1.37':1.37,'3.4':3.4,'7':7,'1.7':1.7,'4':4,'1.1':1.1,'4':4,'9.48':9.48,'1.14':1.14,'1.2':1.2,'1.29':1.29} #Then map to numeric
BXHER2_COPY = {'Unknown':np.NaN,'1':1,'2.09':2.09,'16.8':16.8,'13.63':13.63,'3':3,'1.9':1.9,'22.75':22.75,'2.3':2.3,'2.2':2.2,'2.78':2.78} #Then map to numeric
BXCEP17_COPY = {'Unknown':np.NaN,'3.38':3.38,'1.8':1.8,'2.6':2.6,'2.16':2.16} #Then map to numeric
```

```python
data = pd.read_csv('data_clinical_sample.txt', delimiter='\t',skiprows=4)
data.drop(['SAMPLE_TIMEPOINT','BX_HISTOLOGY','BX_NOTES'], axis=1,inplace=True) 

#Redundant variables =['BX_PERCENT_ER','BX_PERCENT_PR']

# Convert percent to float
def p2f(x):
    return float(x.strip('%'))/100

data.head()
```

```python
# Map columns based on dictionaries defined above

data['ONCOTREE_CODE'].replace(ONCOTREE_CODE, inplace=True)
data['CALC_TREATMENT_NAIVE'].replace(CALC_TREATMENT_NAIVE, inplace=True)
data['CALC_MET_SETTING'].replace(CALC_MET_SETTING, inplace=True)
data['BX_DCIS'].replace(BX_DCIS, inplace=True)
data['BX_LCIS'].replace(BX_LCIS, inplace=True)
data['BX_GRADE'].replace(BX_GRADE, inplace=True)
data['BX_ER'].replace(BX_ER, inplace=True)
data['BX_PR'].replace(BX_PR, inplace=True)
data['BX_HER2OVERALL'].replace(BX_HER2OVERALL, inplace=True)
data['BX_HER2IHC'].replace(BX_HER2IHC, inplace=True)
data['BX_HER2FISH'].replace(BX_HER2FISH, inplace=True)

data['BXHER2_RATIO'].replace(BXHER2_RATIO, inplace=True)
data['BXHER2_COPY'].replace(BXHER2_COPY, inplace=True)
data['BXCEP17_COPY'].replace(BXCEP17_COPY, inplace=True)
data[['BXHER2_RATIO','BXHER2_COPY','BXCEP17_COPY']] = data[['BXHER2_RATIO','BXHER2_COPY','BXCEP17_COPY']].apply(pd.to_numeric, errors='coerce')

data.to_csv('brca_mbcproject_wagle_2017/data_clinical_sample_clean.txt', sep='\t')
```

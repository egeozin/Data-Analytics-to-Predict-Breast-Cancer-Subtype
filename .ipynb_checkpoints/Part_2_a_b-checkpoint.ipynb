```python
import pandas as pd
import math
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.style as style
import matplotlib.patches as patches
from matplotlib.collections import PatchCollection
import csv
from sklearn.decomposition import TruncatedSVD
from sklearn import decomposition
from sklearn.linear_model import LogisticRegression
from scipy.sparse import csr_matrix

style.use('fivethirtyeight')
```

```python
#-------- Part 2 (a) ---------#
```

```python
X = np.zeros((272, 1065024))
with open('Trapnell.csv', 'rt') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    for i, row in enumerate(reader):
        X[i] = row
```

```python
svd = TruncatedSVD(n_components=200, n_iter=7, random_state=42)
```

```python
X_reduced = svd.fit_transform(X)
print(svd.explained_variance_ratio_.sum())
```

```python
plt.scatter(X_reduced[:,0], X_reduced[:,1])
plt.axis([0, 0.07, -0.03, 0.04])
plt.tick_params(axis = 'both', which = 'major', labelsize = 10)
plt.tick_params(axis='x', colors='#263238')
plt.tick_params(axis='y', colors='#263238')
plt.gca().set_aspect('equal', adjustable='box')
plt.axhline(y = -0.029, color = 'black', linewidth = 1.0, alpha = .5)
plt.show()
```

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
 
# k means determine k
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(X_reduced)
    kmeanModel.fit(X_reduced)
    distortions.append(sum(np.min(cdist(X_reduced, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / X_reduced.shape[0])
 
# Plot the elbow
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distance')
plt.title('Number of Clusters')
plt.tick_params(axis = 'both', which = 'major', labelsize = 10)
plt.tick_params(axis='x', colors='#263238')
plt.tick_params(axis='y', colors='#263238')
plt.axhline(y = 0.01, color = 'black', linewidth = 1.0, alpha = .5)
plt.show()
```

```python
n_clusters = 2
kmeanModel = KMeans(n_clusters=n_clusters).fit(X_reduced)
kmeanModel.fit(X_reduced)
labels = kmeanModel.labels_
plt.scatter(X_reduced[:,0], X_reduced[:,1], c=labels.astype(np.float))
plt.axis([0, 0.07, -0.03, 0.04])
plt.tick_params(axis = 'both', which = 'major', labelsize = 10)
plt.tick_params(axis='x', colors='#263238')
plt.tick_params(axis='y', colors='#263238')
plt.axhline(y = -0.029, color = 'black', linewidth = 1.0, alpha = .5)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

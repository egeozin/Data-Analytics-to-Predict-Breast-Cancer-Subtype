```python
gene_info_f = 'Gene_Info.csv'
gene_interest_f = 'genes_interest.csv'
info = pd.read_csv(gene_info_f, sep=',', )
interest = pd.read_csv(gene_interest_f, sep=',')
info = info[info.chromosome_name.str.contains("CHR") == False]
info.drop(['entrezgene', 'ensembl_gene_id', 'band', 'Unnamed: 0'], axis=1, inplace=True)
```

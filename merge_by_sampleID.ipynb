```python
import numpy as np
import pandas as pd
from IPython.core.display import display, HTML
```

```python
def merge_dataframes(data_frames, index_col='PATIENT_ID'):
    dfs = []
    for data in data_frames:
        data = pd.read_csv(data)
        data.set_index(index_col, inplace=True)
        display(data.head())
        # TODO: Make sample_id the index 
        dfs.append(data)
    
    result = pd.concat(dfs, axis=1)
    
    return result
```

```python
data_frames = ['imputed_patient_clean_with_therapies_1.csv', 'imputed_sample_clean_1.csv']
```
